```python
# Rating data merge inside the books dataset
rating.merge(book,on="ISBN").shape
```

```python
# after merging rating data i create new object "r_name"
r_name=rating.merge(book,on="ISBN")
r_name.head(2)
```

```python
# books dataset of columns "Book-Title" and  "book-rating" both data grouped by groupby function and count the total data
num_rating=r_name.groupby('Book-Title').count()[["Book-Rating"]].reset_index()
num_rating
```

```python
# renaming of columns name 
num_rating.rename(columns={"Book-Rating":"num_rating"},inplace=True)
num_rating
```

```python
# grouping the data of "Book-Title" and "Book-Rating.
# and calculate the "mean" values of "ave_rating" data

ave_rating=r_name.groupby('Book-Title').mean()[["Book-Rating"]].reset_index()

# create new variable which name is ave-rating, and renaming the columns name of "book-rating"
ave_rating.rename(columns={"Book-Rating":"ave_rating"},inplace=True)
ave_rating
```

```python
# define new variable and merge the colmns name of ave-rating in book dataset
popular = num_rating.merge(ave_rating,on="Book-Title")
popular
```

```python
# I select the greter than 250 rating value of books and sort the value of ave-rating
popular=popular[popular["num_rating"]>=250].sort_values('ave_rating',ascending=False).head(5)
popular
```

```python
#  merge the data which sequence is "Book-Title","Book-Author","Image-URL-L","num_rating","ave_rating", after that drop_the_duplictes values

popular=popular.merge(book,on ="Book-Title").drop_duplicates("Book-Title")[["Book-Title","Book-Author","Image-URL-L","num_rating","ave_rating"]]
```

# colabrate based Recomendation system

```python
r_name.head(2)  # r_name inside books of rating its merge data
```

```python
# inside rating data grouping the columns of "User-Id" and "Book-Rating" after that create new x variable
x= r_name.groupby("User-ID").count()["Book-Rating"]
x
```

```python
# after creating x variable i want to select grater than 200 of rating values
x =r_name.groupby("User-ID").count()["Book-Rating"] >200

# after that i iterating the values of x'index and store the new variable which name is "eduser" it's means eduacated user,
#which readed of 200 above books and give the rating of book after reading
eduser = x[x].index
eduser
```

```python
# after that i filtered the "user-ID" those are available in eduser [ it's means >200 of rating user-id persons]
frating=r_name[r_name["User-ID"].isin(eduser)]
frating.head(2)
```

```python
# again filtering the book rating which book rating is >50 and stored in y variable 
y=frating.groupby("Book-Title").count()["Book-Rating"]>=50
y
```

```python
# after stored in y, iteratting the y values in famousebook variable 
famousbook=y[y].index
famousbook.shape
```

```python
# frating means filtered-rating of book which values 200 rating and i checkin the famousbook of book name, 
# which is available or not in "Book-Title" after cheking droped the duplicates values from final dataset

final=frating[frating["Book-Title"].isin(famousbook)]
final.drop_duplicates()
final
```

```python
# DataFrame.pivot_table(index=None, columns=None, values=None), Create a spreadsheet-style pivot table as a DataFrame.
# The levels in the pivot table will be stored in MultiIndex objects (hierarchical indexes) on the index and columns of the result DataFrame.

pt=final.pivot_table(index="Book-Title",columns='User-ID',values="Book-Rating")
pt.fillna(0,inplace=True)
```

```python
pt.shape
```

```python
# The distances between the row vectors of X and the row vectors of Y can be evaluated using pairwise_distances. 
# If Y is omitted the pairwise distances of the row vectors of X are calculated. Similarly, pairwise.pairwise_kernels can be used 
# to calculate the kernel between X and Y using different kernel functions.

# Compute cosine similarity between samples in X and Y. Cosine similarity, or the cosine kernel, computes similarity as the 
# normalized dot product of X and Y:

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
# apply 2 empty list(data,item), inside item=[] storing the book-title,book-author and book's image url then append the data to empty data=[]
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data= []
    for i in similar_items:
        item=[]
        temp = book[book["Book-Title"]==pt.index[i[0]]]
        item.extend(list(temp.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
# append use for 3 -dimensional array
# extend use for 2 -dimensional array
```
